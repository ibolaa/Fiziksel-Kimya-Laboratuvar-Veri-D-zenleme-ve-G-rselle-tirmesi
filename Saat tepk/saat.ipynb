```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#data
at = [15.2,14.6,14,14.9,14.2,12.9,11.7]
t = [3,6,9,12,15,18,21]
```

```python

#data

data = (t,at)
data = pd.DataFrame(data)
data.index = ['a','t']  # type: ignore
data # type: ignore


#plot a linear regression graph
sns.regplot(t,at, color="blue", marker="+",scatter_kws={"s": 30},    line_kws={"linewidth": 0.0005}, ci=None,truncate=True,order=1,robust=True) # type: ignore

plt.xlabel("t")
plt.ylabel("at")
plt.ylim(11.5,15.5)
plt.xlim(2.5,25)
plt.title("t / at")
plt.grid(   axis="both",    which="both",    color="black",    linestyle="--",    linewidth=0.5)
plt.savefig("sakkaroz.png")
plt.show()






```
