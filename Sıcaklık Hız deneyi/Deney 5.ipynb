```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
k = lambda x : 1/x
y = (np.array([3.53,3.35,3.25,3.16]) ) # K
y = k(y)
x = np.array([-3.474,-3.101,-2.865,-2.293])
data = (y, x)
data = pd.DataFrame(data)
data.index = ["y", "x"] # type: ignore
data.columns = ["1", "2", "3", "4"] # type: ignore
data = data.transpose()
data # type: ignore
```

```python
sns.regplot(x=y,y=x,color="blue", marker="+",scatter_kws={"s": 30},line_kws={"linewidth": 0.5}, ci=None,truncate=True,order=1)
plt.xlabel("T (K.10⁻³)")
plt.ylim(-2,-4)
plt.ylabel("1/Δt (s)")


plt.title("Deney 5 - 1/Δt vs T Lineer Regresyon Grafiği")
plt.tight_layout()
plt.grid(   axis="both",    which="both",    color="black",    linestyle="-",    linewidth=0.5)
plt.show()
plt.savefig("Deney 5 - Δt vs T Lineer Regresyon Grafiği.png")
```
