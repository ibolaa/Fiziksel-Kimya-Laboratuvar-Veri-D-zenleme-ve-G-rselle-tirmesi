```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
T = [20,25,30,35]
F = [10.78,11,7.7,7]
yzy = [60.02,61.24,42.90,39.01]
vui = [459.4156,468.7914,328.1540,298.3218]
yzy = yzy 
```

```python
#plot a linear regression graph
sns.regplot(T,F, color="blue", marker="+",scatter_kws={"s": 30},    line_kws={"linewidth": 0.5}, ci=None,truncate=True,order=1,robust=True) # type: ignore

plt.ylabel("Force")
plt.xlabel("Celcius")
plt.xlim(18,36)
plt.ylim(6,12)
plt.title("T/F")
plt.grid(   axis="both",    which="both",    color="black",    linestyle="--",    linewidth=0.5)
plt.savefig("Du nouy regplot.png")
plt.show()
```

```python
from scipy import stats
slope, intercept, r_value, p_value, std_err = stats.linregress(T,yzy)
print("slope: %f    intercept: %f" % (slope, intercept))
print("r-squared: %f" % r_value**2)
print("p_value: %f" % p_value)
print("std_err: %f" % std_err)

x = range(0,100)
y = slope*x + intercept

plt.plot(x,y)
plt.ylabel("yzy")
plt.xlabel("Celicus")
plt.xlim(0,60)
plt.ylim(0,100)
plt.title("T/yzy")
plt.grid(   axis="both",    which="both",    color="black",    linestyle="--",    linewidth=0.5)

```
