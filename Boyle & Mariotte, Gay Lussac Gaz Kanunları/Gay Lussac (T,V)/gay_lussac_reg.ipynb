```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#data
x = np.array([26,31,37,43,48])
y = np.array([0.048,0.049,0.050,0.051,0.052])
data =(x,y)
data = pd.DataFrame(data)
data.index = ["x","y"]   # type: ignore
data.columns = ["1","2","3","4","5"] # type: ignore
data
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
#regplot
sns.regplot(x=x, y=y, data=data, ax=ax, color="blue", marker="+", scatter_kws={"s": 100}, line_kws={"linewidth": 0.5},ci=None, truncate=True,)

#marker="+",scatter_kws={"s": 30},    line_kws={"linewidth": 0.5}, ci=None,truncate=True,order=1,robust=True
ax.set_title(label="Gay Lussac Lineer Regresyon Grafiği")
ax.set_xlabel(xlabel="T (°C)")
ax.set_ylabel(ylabel="P (atm)")
ax.grid(   axis="both",    which="both",    color="black",    linestyle="-",    linewidth=0.5) # type: ignore
ax.set_xlim(23,50)
ax.set_ylim(0.046,0.053)
plt.show()
```
