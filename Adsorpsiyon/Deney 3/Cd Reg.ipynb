```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
A = [1,3,5,8]
C = [0.138,0.419,0.807,1.105]
```

```python

#data

data = (A,C)
data = pd.DataFrame(data)
data.index = ['A','C']  # type: ignore
data # type: ignore

# Deney 1 - Boyle Mariotte

#plot a linear regression graph
sns.regplot(x=A,y=C,color="blue", marker="+",scatter_kws={"s": 30},line_kws={"linewidth": 0.5}, ci=None,truncate=True,order=1) # type: ignore
plt.xlabel("A")
plt.ylabel("C")


plt.title("A / C REG")
plt.grid(   axis="both",    which="both",    color="black",    linestyle="-",    linewidth=0.5)
plt.savefig("A_C REG.png")
plt.show()





```
